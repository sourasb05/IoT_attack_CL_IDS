```python
import json
import matplotlib.pyplot as plt

# Load data
with open('/proj/sourasb-220503/IoT_attack_CL_IDS/results/1_experiment_results_LSTM_WCL_toggle.json', 'r') as f:
    data1 = json.load(f)
with open('/proj/sourasb-220503/IoT_attack_CL_IDS/results/1_experiment_results_LSTM_LwF_toggle_alpha_1.0_T_4.0.json', 'r') as f:
    data2 = json.load(f)
with open('/proj/sourasb-220503/IoT_attack_CL_IDS/results/1_experiment_results_LSTM_GR_toggle.json', 'r') as f:
    data3 = json.load(f)
with open('/proj/sourasb-220503/IoT_attack_CL_IDS/results/1_experiment_results_LSTM_EWC_toggle.json', 'r') as f:
    data4 = json.load(f)
with open('/proj/sourasb-220503/IoT_attack_CL_IDS/results/1_experiment_results_LSTM_Replay_REPLAY_toggle.json', 'r') as f:
    data5 = json.load(f)

# OkabeIto color-blindfriendly palette
COLORS = {
    "W/O CL": "#0072B2",   # blue
    "LwF":    "#E69F00",   # orange
    "GR":     "#D55E00",   # vermilion
    "EWC":    "#CC79A7",   # reddish-purple
    "Replay": "#009E73",   # bluish-green
}

plt.figure(figsize=(8,6), dpi=150)

plt.plot(data1["Utility"], label="W/O CL", color=COLORS["W/O CL"],
         lw=2.5, ls="-", marker="o", markersize=5)
plt.plot(data2["Utility"], label="LwF", color=COLORS["LwF"],
         lw=2.5, ls="--", marker="^", markersize=5)
plt.plot(data3["Utility"], label="GR", color=COLORS["GR"],
         lw=2.5, ls="-.", marker="s", markersize=5)
plt.plot(data4["Utility"], label="EWC", color=COLORS["EWC"],
         lw=2.5, ls=":", marker="D", markersize=5)
plt.plot(data5["Utility_WP_2"], label="Replay", color=COLORS["Replay"],
         lw=2.5, ls="-", marker="*", markersize=6)

plt.xlabel("Domain Index", fontsize=16)
plt.ylabel("Utility", fontsize=16)
plt.title("Utility Values Comparison (Toggle)", fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True, alpha=0.3)

plt.legend(title="Method", fontsize=12, title_fontsize=12,
           frameon=True, edgecolor="black", loc="upper right")

plt.tight_layout()
plt.savefig("/proj/sourasb-220503/IoT_attack_CL_IDS/results/plots/utility_toggle_comparison_cb.png")
plt.show()
```
